Pregunta 2:

1. Crea un DataFrame df_filtrado con todas las filas de df_diabetes donde Glucose > 120.
2. Agrega a df_filtrado una columna BMI_Category que clasifique el BMI así:

    * "Bajo" si BMI < 18.5
    * "Normal" si 18.5 ≤ BMI < 25
    * "Sobrepeso" si 25 ≤ BMI < 30
    * "Obeso" si BMI ≥ 30

3. Muestra las primeras 5 filas de df_filtrado con la nueva columna incluida.

```python
#Creating new dataframe
df_filtrado = df_diabetes[df_diabetes['Glucose'] > 120]
```

```python
#Adding the BMI column according to the categories given
df_filtrado = df_filtrado.copy()
df_filtrado['BMI_Category'] = pd.cut(df_filtrado['BMI'], bins=[0, 18.5, 25, 30, float('inf')], labels=['Bajo', 'Normal', 'Sobrepeso', 'Obeso'])
```

```python
df_filtrado.head()
```

Pregunta 3:
1. Carga age_groups.csv en un DataFrame df_age_groups.
2. Haz un merge de df_diabetes con df_age_groups por la columna Age. El resultado se llamará df_combinado. Debe ser un left merge para mantener todas las filas de df_diabetes.
3. Muestra las 5 primeras fila

```python
#Uploading age_groups.csv to a dataframe
df_age_groups = pd.read_csv('age_groups.csv')
```

```python
#Merging diabetes dataser and age_groups dataset
df_combinado = pd.merge(df_diabetes, df_age_groups, on='Age', how='left')
```

```python
#printing
df_combinado.head()
```

Pregunta 4:
1. Agrupa df_combinado por AgeGroup.
2. Calcula el promedio de Glucose y el promedio de BMI para cada grupo de edad.
3. Ordena el resultado por el promedio de Glucose de mayor a menor.
4. Muestra el DataFrame resultante.

```python
#Grouping df_combinado by AgeGroup
df_grouped = df_combinado.groupby('AgeGroup')
```

```python
#Mean of Glucose and BMi by agegroup
df_grouped = df_combinado.groupby('AgeGroup')[['Glucose', 'BMI']].mean()
```

```python
#Ascending order of Glucose mean
df_grouped = df_grouped.sort_values(by='Glucose', ascending=False)
```

```python
#printing
df_grouped.head()
```

  Pregunta 5:
1. A partir de df_combinado, filtra las filas donde AgeGroup = "Adulto" y Outcome = 1 (casos positivos
de diabetes en adultos).
2. De este subconjunto, selecciona los 5 registros con el mayor valor de BMI.
3. Muestra únicamente las columnas Age, Glucose, BMI, DiabetesPedigreeFunction, ordenadas por BMI de mayor a menor.

```python
#Filtering as requested
df_filtered = df_combinado[(df_combinado['AgeGroup'] == 'Adulto') & (df_combinado['Outcome'] == 1)]
```

```python
#Showing the top5 BMI of df_filtered
df_top5 = df_filtered.nlargest(5, 'BMI')
```

```python
#Output as requested
df_top5[['Age', 'Glucose', 'BMI', 'DiabetesPedigreeFunction']].sort_values(by='BMI', ascending=False)
```

Pregunta 6:
1. Usando df_combinado, crea una tabla pivote que tenga:
    * Índice: AgeGroup
    * Columnas: Outcome
    * Valores: promedio de Glucose y BMI
2. La pivot table debe mostrar el promedio de Glucose y BMI por combinación de AgeGroup y Outcome.
3. De la tabla resultante, filtra y muestra únicamente las filas (AgeGroups) donde el promedio de BMI para
cualquiera de los valores de Outcome sea mayor a 30.

```python
#Creating a pivot table
df_pivot = pd.pivot_table(
    df_combinado,
    index='AgeGroup',
    columns='Outcome',
    values=['Glucose', 'BMI'],
    aggfunc='mean',
    observed=False
)
```

```python
filtro_bmi_pivot = df_pivot[(df_pivot[('BMI', 0)] > 30) | (df_pivot[('BMI', 1)] > 30)]
```
